# **IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **CREATING DATAFRAME AND SAVING TO CSV**




```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

# **VISUALIZING THE DATA**

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.xlabel('Hours of Study (X)')
plt.ylabel('Grades (Y)')
plt.title('') #Scatter Plot of Data
plt.show()
```

# **PREPARING DATA FOR MODELING**

```python
X = data[['Hours_of_Study']] #Independent Variable
Y = data[['Grades']] #Dependent Variable
```

```python
model = LinearRegression()

model.fit(X,Y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color='black')
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

```python
new_Hours_of_Study = [[np.random.randint(0,24)]]
predicted_Grades = model.predict(new_Hours_of_Study)

print('Hours of Study: ',new_Hours_of_Study)
print('Predicted Grades: ',predicted_Grades)
```
